## 12.2 두 개의 독립표본
* 두 모집단으로부터 추출된 표본과 통계량
    * 모집단 1 : 
$ \\
평균 : \mu_{1} \\
표준편차 : \sigma_{1} \\
표본 : X_{1},X_{2},X_{2},...,X_{n1} \\
표본의 수 : n_{1} \\
표본평균 : \bar{X} = \frac{1}{n_{1}} \sum\limits_{i=1}^{n_{1}}{X_{i}}\\
표본분산 : S^{2}_{1} = \frac{1}{(n_{1}-1)}\sum\limits_{i=1}^{n_{1}}{(X_{i}-\bar{X})^2}
$ 
    * 모집단 2 : 
$ \\
평균 : \mu_{2} \\
표준편차 : \sigma_{2} \\
표본 : Y_{1},Y_{2},Y_{2},...,Y_{n_{2}} \\
표본의 수 : n_{2} \\
표본평균 : \bar{Y} = \frac{1}{n_{2}} \sum\limits_{i=1}^{n_{2}}{Y_{i}}\\
표본분산 : S^{2}_{2} = \frac{1}{(n_{2}-1)}\sum\limits_{i=1}^{n_{2}}{(Y_{i}-\bar{Y})^2}
$

* 모평균의 차($ \mu_{1}- \mu_{2} $) 에 대한 추론  
$ 
   \bar{X} \sim N(\mu_{1},\frac{\sigma_{1}^{2}}{n_{1}}) \\
   \bar{Y} \sim N(\mu_{2},\frac{\sigma_{2}^{2}}{n_{2}})  \\
$
    * 모분산을 알 때   
$    
   (\bar{X}-\bar{Y}) \sim N(\mu_{1}-\mu_{2},\frac{\sigma_{1}^{2}}{n_{1}}+\frac{\sigma_{2}^{2}}{n_{2}})
$
    * 모분산을 모를 때   
$\\
   Z = \frac{ (\bar{X}-\bar{Y}) - (\mu_{1}-\mu_{2})}{\sqrt{ s^{2}_{1}/n_{1} + s^{2}_{2}/n_{2}}} \sim N(0,1)
$
    * $(\mu_{1}-\mu_{2})$에 대한 100(1-$\alpha$)% 신뢰구간 (표본의 크기가 클 때)    
$
  (\bar{X}-\bar{Y}) \pm z_{\alpha/2}\sqrt{ s^{2}_{1}/n_{1} + s^{2}_{2}/n_{2}}
$
    * H0: $ \mu_{1} - \mu_{2} = \delta_{0} $에 대한 검정 (표본의 크기가 클 때)   
        * 검정통계량    
$    
Z = \frac{ (\bar{X}-\bar{Y}) - \delta_{0}}{\sqrt{ s^{2}_{1}/n_{1} + s^{2}_{2}/n_{2}}} \sim N(1,0) 
$
        * 각 대립가설(H1)의 유의수준 $\alpha$ 에 대한 기각역  
           * H1: $ \mu_{1} - \mu_{2} < \delta_{0}$ 일 때,    R : $ Z \leq -z_{\alpha} $
           * H1: $ \mu_{1} - \mu_{2} > \delta_{0}$ 일 때,    R : $ Z \geq z_{\alpha} $
           * H1: $ \mu_{1} - \mu_{2} \neq \delta_{0}$ 일 때,    R : $ |Z| \geq z_{\alpha / 2} $


* 모평균의 차에 대한 추론 (표본의 크기가 작을 때)
    * 필요한 가정
        * 두 모집단의 모두 정규분포를 따른다
        * 두 모집단의 표준편차가 일치한다 ($\sigma_{1} =  \sigma_{2} = \sigma$)
    * 공통분산($\sigma^{2}$)에 대한 합동 추정량   
$
    S^{2}_{p} = \frac{\sum\limits_{i=1}^{n_{1}}{(X_{i}-\bar{X})^2} + \sum\limits_{i=1}^{n_{2}}{(Y_{i}-\bar{Y})^2}}{n_{1}+n_{2}-2}
$
    * 통계량 t   
$
    t = \frac{ (\bar{X}-\bar{Y}) - (\mu_{1}-\mu_{2}) }{ S_{p}\sqrt{1/n_{1}+1/n_{2}}} \sim t(n_{1}+n_{2}-2)
$
    * $(\mu_{1}-\mu_{2})$에 대한 100(1-$\alpha$)% 신뢰구간 (표본의 크기가 작고 모분표분편차가 같을 때) 
$
  (\bar{X}-\bar{Y}) \pm t_{\alpha/2}S_{p}\sqrt{ 1/n_{1} + 1/n_{2}}
$
    * H0: $ \mu_{1} - \mu_{2} = \delta_{0} $에 대한 검정 (표본의 크기가 작고 모분표분편차가 같을 때) 
        * 검정통계량   
$    t = \frac{ (\bar{X}-\bar{Y}) - \delta_{0}}{S_{p}\sqrt{1/n_{1} + 1/n_{2}}} \sim t(n_{1}+n_{2}-2)
$
        * 각 대립가설(H1)의 유의수준 $\alpha$ 에 대한 기각역  
           * H1: $ \mu_{1} - \mu_{2} < \delta_{0}$ 일 때,    R : $ t \leq -t_{\alpha}(n_{1}+n_{2}-2) $
           * H1: $ \mu_{1} - \mu_{2} > \delta_{0}$ 일 때,    R : $ t \geq t_{\alpha}(n_{1}+n_{2}-2) $
           * H1: $ \mu_{1} - \mu_{2} \neq \delta_{0}$ 일 때,    R : $ |t| \geq t_{2/\alpha}(n_{1}+n_{2}-2) $

    * $(\mu_{1}-\mu_{2})$에 대한 100(1-$\alpha$)% 신뢰구간 (표본의 크기가 작고 모분표분편차가 다를 때)   
$
    t^{*} = \frac{ (\bar{X}-\bar{Y}) - \delta_{0}}{\sqrt{s_{1}^{2}/n_{1} + s_{2}^{2}/n_{2} } } \sim t(min(n_{1}-1,n_{2}-1)) \\
    (\bar{X}-\bar{Y}) \pm t^{*}_{\alpha/2}\sqrt{ s^{2}_{1}/n_{1} + s^{2}_{2}/n_{2}}
$    
    * H0: $ \mu_{1} - \mu_{2} = \delta_{0} $에 대한 검정 (표본의 크기가 작고 모분표분편차가 다를 때)
        * 검정통계량   
$ 
t^{*} = \frac{ (\bar{X}-\bar{Y}) - \delta_{0}}{\sqrt{s_{1}^{2}/n_{1} + s_{2}^{2}/n_{2} } } \sim t(min(n_{1}-1,n_{2}-1))
$    
        * 각 대립가설(H1)의 유의수준 $\alpha$ 에 대한 기각역  
       * H1: $ \mu_{1} - \mu_{2} < \delta_{0}$ 일 때,    R : $ t \leq -t^{*}_{\alpha} $
       * H1: $ \mu_{1} - \mu_{2} > \delta_{0}$ 일 때,    R : $ y \geq t^{*}_{\alpha} $
       * H1: $ \mu_{1} - \mu_{2} \neq \delta_{0}$ 일 때,    R : $ |t| \geq t^{*}_{\alpha / 2} $

#### 정리
* 두개의 독립표본에 대하여
    * 표본의 크기가 클 때
        * 모표준편차를 알 때
            * 검정통계량  
$    
Z = \frac{ (\bar{X}-\bar{Y}) - (\mu_{1}-\mu_{2})}{\sqrt{ \sigma^{2}_{1}/n_{1} + \sigma^{2}_{2}/n_{2}}} \sim N(1,0) 
$            
        * 모표준편차를 모를 때
            * 검정통계량    
$    
Z = \frac{ (\bar{X}-\bar{Y}) - (\mu_{1}-\mu_{2})}{\sqrt{ s^{2}_{1}/n_{1} + s^{2}_{2}/n_{2}}} \sim N(1,0) 
$        
    * 표본의 크기가 작을 때 ( 두모집단이 정규분포를 따르는 조건에서)
        * 모표준편차가 같을 때
            * 공통분산 합동추정량   
$
    S^{2}_{p} = \frac{\sum\limits_{i=1}^{n_{1}}{(X_{i}-\bar{X})^2} + \sum\limits_{i=1}^{n_{2}}{(Y_{i}-\bar{Y})^2}}{n_{1}+n_{2}-2}
$
            * 검정통계량   
$
    t = \frac{ (\bar{X}-\bar{Y}) - (\mu_{1}-\mu_{2}) }{ S_{p}\sqrt{1/n_{1}+1/n_{2}}} \sim t(n_{1}+n_{2}-2)
$            
        * 모표준편차가 다를 때
            * 검정통계량   
$ 
t^{*} = \frac{ (\bar{X}-\bar{Y}) - (\mu_{1}-\mu_{2})}{\sqrt{s_{1}^{2}/n_{1} + s_{2}^{2}/n_{2} } } \sim t(min(n_{1}-1,n_{2}-1))
$

## 12.3 짝비교
* 짝비교 자료형태
    * 샘플개수 : n
    * $D_{i} = X_{i}-Y{i}$
    * $\bar{D} = \frac{1}{n} \sum\limits_{i=1}^{n}{D_{i}}$
    * $S_{D}^2 = \frac{1}{(n-1)}\sum\limits_{i=1}^{n}{(D_{i}-\bar{D})^2}$
    
* 평균차 $\delta$에 대한 추론
    * 표본의 크기가 클 때  
$ 
Z= \frac{\bar{D}-\delta}{S_{D} / \sqrt{n}} \sim N(1,0)
$
    * 표본의 크기가 작을 때   
$ 
t= \frac{\bar{D}-\delta}{S_{D} / \sqrt{n}} \sim t(n-1)
$
    
        
    

## 12.4 두 모비율의 차에 대한 추론
* 두 모집단에서 모비율을 비교(두 표본은 서로 독립) 
    * 모집단 1 
        * 모비율 : $p_{1}$
        * 표본크기 : n1
        * 특성이 A인 표본수 : X
        * 표본비율 : $\hat{p}_{1} = \frac{X}{n1}$
        * $X \sim N(n_{1}p_{1},n_{1}p_{1}(1-p_{1}))$
        * $\hat{p}_{1} = \frac{X}{n_{1}} \sim N(p_{1},\frac{p_{1}(1-p_{1})}{n_{1}})$ 
    * 모집단 2 
        * 모비율 : $p_{2}$
        * 표본크기 : n2
        * 특성이 A인 표본수 : Y
        * 표본비율 : $\hat{p}_{2} = \frac{Y}{n2}$  
        * $Y \sim N(n_{2}p_{2},n_{2}p_{2}(1-p_{2}))$
        * $\hat{p}_{2} = \frac{Y}{n_{2}} \sim N(p_{2},\frac{p_{2}(1-p_{2})}{n_{2}})$ 
    * 표본비율의 차의 분포  
$    
    (\hat{p}_{1}-\hat{p}_{2}) \sim N(p_{1}-p_{2},\frac{p_{1}(1-p_{1})}{n_{1}}+\frac{p_{2}(1-p_{2})}{n_{2}}) \\
    Z = \frac{ (\hat{p}_{1}-\hat{p}_{2}) - (p_{1}-p_{2})}{ \sqrt{\frac{p_{1}(1-p_{1})}{n_{1}}+\frac{p_{2}(1-p_{2})}{n_{2}}} } \sim N(1,0)
$        
    * 모비율의 차에 대한 신뢰구간(표본의 크기가 클 때)   
$
    (\hat{p}_{1}-\hat{p}_{2}) \pm z_{2/\alpha}\sqrt{\frac{\hat{p}_{1}(1-\hat{p}_{1})}{n_{1}}+\frac{\hat{p}_{2}(1-\hat{p}_{2})}{n_{2}}} 
$
        * 모비율 $p_{1},p_{2}$는 미지수 이므로 대신 $\hat{p}_{1},\hat{p}_{2}$ 사용    
    * H0: $p_{1}=p_{2}$ 에 대한 검정(표본의 크기가 클 때)   
$
    \hat{p} = \frac{X+Y}{n_{1}+n_{2}} (공통비율) \\
    Z = \frac{ (\hat{p}_{1}-\hat{p}_{2})}{ \sqrt{\hat{p}(1-\hat{p})}\sqrt{1/n_{1}+1/n_{2}}  } \sim N(1,0)
$
        * 각 대립가설(H1)의 유의수준 $\alpha$ 에 대한 기각역  
           * H1: $ \mu_{1} - \mu_{2} < \delta_{0}$ 일 때,    R : $ Z \leq -z_{\alpha} $
           * H1: $ \mu_{1} - \mu_{2} > \delta_{0}$ 일 때,    R : $ Z \geq z_{\alpha} $
           * H1: $ \mu_{1} - \mu_{2} \neq \delta_{0}$ 일 때,    R : $ |Z| \geq z_{\alpha / 2} $

#### H0
#### 두 신발 원재료 A, B는 닳는 정도가 같다 
####  (mu1 = mu2, ie. difference = 0)
#### H1 
####  두 신발 원재료 A, B는 닳는 정도가 다르다
####  (mu1 != mu2, ie. difference != 0) => 양측검정

##### 예제 데이터
library(MASS)
str(shoes)
##### t.test 사용해보자
t_test = t.test(shoes$A, shoes$B,
       alternative = c("two.sided"), 
       paired = TRUE, 
       conf.level = 0.95)
print(t_test)
##### 직접구해보자!
D = shoes$A - shoes$B
n = length(D)
mean_D = mean(D) #표본평균
sd_D = sd(D)     #표본표준편차
t = (mean_D - 0) /(sd_D/sqrt(n)) 
print(t)
alpha = 0.05
t_half_alpha  = qt(alpha/2, n-1, lower.tail = FALSE, log.p = FALSE)
print(t_half_alpha)
t_interval = c(mean_D-t_half_alpha*sd_D/sqrt(n),mean_D+t_half_alpha*sd_D/sqrt(n))
print(t_interval)
pval = 2*pt(t, n-1, lower.tail = TRUE, log.p = FALSE)
print(pval)
