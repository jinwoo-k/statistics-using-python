```python
##표본 평균 저장

import numpy as np
m = []
np.random.seed(1234)
for i in range(100):
    sample=np.random.randint(0, 10, size = 5)
    m.append(np.mean(sample))
m = np.array(m)

print(m)
```

```python
##히스토그램

import matplotlib.pyplot as plt
plt.hist(m, bins=9)
plt.xlabel('m')
plt.ylabel('Frequency')
plt.title('Historam of m')
plt.show()
```

```python
#정규 확률 그림
import matplotlib.pyplot as plt
import statsmodels.api as sm

sm.qqplot(m,line='s')
plt.title("Normal Q-Q plot")
```
