## 6.3 이산확률변수와 확률분포
**확률분포(probability distribution)** : 확률변수가 갖는 __값__ 들과 그에 대응하는 **확률값** 을 나타내는 것으로 나열된 표 또는 수식으로 표현된다. 보통은 확률변수 *X*의 분포라고 한다.

<br>

이산확률변수의 확률분포

|*X*가 취하는 값(x)|확률 *f*(x)|
|-------------|--------|
|$x_{1}$|$f(x_{1})$|
|$x_{2}$|$f(x_{2})$|
|...|...|
|$x_{k}$|$f(x_{k})$|
|합계|1|

$f(x)$ 를 **확률함수** 또는 __확률질량함수__(PMF: probability mass function)이라 한다.

<br> 

이산확률변수 *X*의 확률변수 $f(x_{i})=P(X=x_{i})$는 다음 조건을 만족해야한다.

1) 모든 $x_{i}$ 값에 대해 $0 <= f(x_{i}) <= 1$

2) $\sum_{모든 x_{i}} f(x_{i}) = 1$

<br>

모집단의 확률분포를 알 수 없는 경우 크기가 큰 표본의 히스토그램을 이용해 확률분포 추정

## 6.4 확률분포의 기대값(평균)과 표준편차
이산확률변수의 평균과 표준편차에 대해 배운다

### 기대값 (Expected Value)
확률변수 X의 기대값(평균) = $ E(X) = µ = \sum(확률변수가 취하는 값 X 그 값을 가질 확률) = \sum x_i f(x_i) $

### 표준편차
$Var(X) = E(X - µ)^2 = \sum (x_i - µ)^2 f(x_i)$

$sd(X) = +\sqrt{Var(X)}$

**분산의 계산식** $Var(x) = E(X^2)-(E(X))^2 = \sum{x_i^2f(x_i)-µ^2}$

## 6.5 두 확률변수의 결합분포
두 확률변수 X 와 Y 의 **결합확률분포**(joint probability distribution)란 *X* 가 취하는 값과 _Y_ 가 취하는 값의 쌍에 대응하는 확률

$f(x_i, y_i) = P[X=x_i, Y=y_i]$

<img src="./img/t6-5.png" />

확률들이 이론적으로 계산되지 않을 경우 상대도수의 수렴치로 확률을 계산한다.

특정 $x_i$ 의 모든 $y_j$(혹은 특정 $y_i$ 의 모든 $x_j$)의 합계를 *주변확률분포*(marginal probability distribution)라 한다.

$$f_x(x_i) = P[X=x_i] = \sum_{j=1}^{n}f(x_i, y_j)$$
$$f_y(y_j) = P[Y=y_j] = \sum_{i=1}^{m}f(x_i, y_j)$$

## 6.6 공분산과 상관계수
공분산은 두개의 확률변수 X와 Y가 서로 어떤 관계를 가지며 변화하는가를 나타내는 척도

X와 Y의 **공분산**(Covariance) : $Cov(X, Y) = E(X-µ_x)(Y-µ_y) = E(XY)-µ_xµ_y $

공분산은 X나 Y의 단위에 영향을 받는다. 영향을 받지 않기 위해서는 상관계수 이용

X와 Y의 **상관계수**(Correlation Coefficient) : $Corr(X, Y) = \frac{Cov(X, Y)}{σ_x σ_y} $
