## 8.3 정규분포의 일반적인 성질 및 확률 계산

* $평균 \mu, 분산 \delta^2 를 갖는 정규분포 : N(\mu, \delta^2) $
* X가 평균 $ \mu로부터 \pm\delta \pm2\delta \pm3\delta$ 사이에 있을 확률
    * $ P[\mu-\delta \le X \le \mu+\delta] = $  68.27%
    * $ P[\mu-2\delta \le X \le \mu+2\delta] = $  95.45%
    * $ P[\mu-3\delta \le X \le \mu+3\delta] = $  99.73%

![정규분포 그래프](./images/Normal_Distribution_PDF.png)

* 중요 확인: $ P[Z \le -z] = P[Z \ge z] = P[Z \gt z] = 1-P[Z \le z]$
* 표준정규확률변수 (standard normal random variable)
    * $확률변수 X 가 N(m, \delta^2) 일때, 표준화된 확률변수 Z=\cfrac{X-\mu}{\delta} 는 평균 0, 분산 1인 N(0,1)$

## 8.4 이항분포의 정규분포근사 
* 이항분포: 성공확률이 p 인 베르누이 실행을 n 번 반복할때 성공의 횟수가 갖는 분포
    * 포아송분포는 n이 아주 크고 p 는 아주 작아서 np = m(상수)가 될때이다.
    * n 이 아주 크고 p 가 0이나 1에 가깝지 않을때는 정규분포에 가까움 (즉, np와 n(1-p) 가 모두 클때)

![정규분포근사](./images/appro_nd.png)
https://soyulsohee-father.tistory.com/20

* 이항분포의 정류분포근사: $X ~Bin(n,p) 이고 np나 n(1-p)가 모두 클 경우 (보통 10이상) 근사적으로 평균이 np 이고 표준편차가 \sqrt{np(1-p)}$ 인 정규분포를 따름

* 연속성 수정 (continuity correction) 
    * $이항 분포에서 [a \le X \le b] 일 확률은 정규분포에서 (a-\cfrac{1}{2} \le X \le b+\cfrac{1}{2}) $

```python
##예제13

import numpy as np
import matplotlib.pyplot as plt

data = np.array([39.3,14.8,6.3,0.9,6.5,
                3.5,8.3,10.0,1.3,7.1,
                6.0,17.1,16.8,0.7,7.9,
                 2.7,26.2,24.3,17.7,3.2,
                 7.4,6.6,5.2,8.3,5.9,
                 3.5,8.3,44.8,8.3,13.4,
                 19.4,19.0,14.1,1.9,12.0,
                 19.7,10.3,3.4,16.7,4.3,
                 1.0,7.6,28.3,26.2,31.7,
                 8.7,18.9,3.4,10.0])

plt.hist(data, bins=5, range=(0,50))
plt.xlabel('data')
plt.ylabel('Density')
plt.title('Historam of data')
plt.show()

sm.qqplot(data,line='s')
plt.title("Normal Q-Q plot")
```
